# 02 - NumPy Operations

In this section, we learn how to:
- Perform mathematical operations on arrays
- Use element-wise operations
- Use aggregation functions (sum, mean, max, min)
- Apply comparisons and Boolean operations


1) Element-wise operations

```python
import numpy as np

a = np.array([10, 20, 30, 40])
b = np.array([1, 2, 3, 4])

print(a + b)
print(a - b)
print(a * b)
print(a / b)
```

2) Scalar operations

```python
print(a + 10)
print(a * 2)
print(a / 5)
```

3) Mathematical functions

```python
print(np.log(a))
print(np.sqrt(a))
print(np.exp(a))
```

4) Aggregations (very important for Data Science)

```python
data = np.array([[1, 2, 3],
                 [4, 5, 6]])

print(data.sum())
print(data.sum(axis=0))  # column-wise
print(data.sum(axis=1))  # row-wise

print(data.mean())
print(data.max())
print(data.min())
print(np.median(data))
```

5) Boolean operations

```python
print(a > 20)
print(a[a > 20])  # filter values
```
